```python
import os
import cv2
import pytesseract
import numpy as np

myconfig = "--psm 11"

def retrieve_images_from_folder(folder_path, query_text):
    matching_images = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            image_path = os.path.join(folder_path, filename)
            image = cv2.imread(image_path)
            gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            text = pytesseract.image_to_string(gray_image, config=myconfig)
            print(text)
            if query_text in text:
                matching_images.append(image)
    
    return matching_images

folder_path = "D:\PD 2"
query_text = 'Good'  # Use '=' instead of '==' for variable assignment

images = retrieve_images_from_folder(folder_path, query_text)

for img in images:
    cv2.imshow('Matching Image', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
